#Visualization to the target population

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x="Customer type",data=dataset)
plt.show()
```

```python
#Exploring relationship between customer type and gender
sns.countplot(x="Customer type",hue="Gender",data=dataset)
plt.show()
```

```python
#Exploring distribution of branch
sns.histplot(x="City",data=dataset)
plt.show()
```
